```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Step 1: Load dataset directly from a website
url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv"
df = pd.read_csv(url)

print("First 5 rows of dataset:")
print(df.head())

# Step 2: Five Different Visualizations

# 1. Histogram (distribution of sepal length)
plt.figure(figsize=(8,5))
sns.histplot(df['sepal_length'], kde=True, bins=20, color="skyblue")
plt.title("Histogram of Sepal Length")
plt.show()

# 2. Scatter Plot (sepal length vs sepal width)
plt.figure(figsize=(8,5))
sns.scatterplot(data=df, x="sepal_length", y="sepal_width", hue="species", palette="Set2")
plt.title("Sepal Length vs Sepal Width by Species")
plt.show()

# 3. Box Plot (distribution of petal length by species)
plt.figure(figsize=(8,5))
sns.boxplot(data=df, x="species", y="petal_length", palette="pastel")
plt.title("Box Plot of Petal Length by Species")
plt.show()

# 4. Pairplot (relationships among all features)
sns.pairplot(df, hue="species", palette="Set1")
plt.suptitle("Pairplot of Iris Dataset", y=1.02)
plt.show()
```

```python
# 5. Heatmap (correlation matrix) - FIXED
plt.figure(figsize=(8,6))
sns.heatmap(df.select_dtypes(include="number").corr(), annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Correlation Heatmap of Iris Features")
plt.show()
```
