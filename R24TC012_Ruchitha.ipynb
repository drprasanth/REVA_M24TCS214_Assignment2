```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler

# Load dataset
df = pd.read_csv("/content/drive/MyDrive/DelayedFlights.csv")

# Step 1: Data Cleaning
df_cleaned = df.drop(columns=["Unnamed: 0"], errors="ignore")  # drop index column if present
df_cleaned = df_cleaned.dropna()  # remove rows with missing values

# Step 2: Normalization (Min-Max Scaling on numerical columns)
numeric_cols = df_cleaned.select_dtypes(include=[np.number]).columns
scaler = MinMaxScaler()
df_cleaned[numeric_cols] = scaler.fit_transform(df_cleaned[numeric_cols])

# Step 3: Mean, Median, Mode
mean_vals = df_cleaned[numeric_cols].mean()
median_vals = df_cleaned[numeric_cols].median()
mode_vals = df_cleaned[numeric_cols].mode().iloc[0]

print("=== Mean (first 10) ===")
print(mean_vals.head(10))

print("\n=== Median (first 10) ===")
print(median_vals.head(10))

print("\n=== Mode (first 10) ===")
print(mode_vals.head(10))

# Step 4: Covariance Matrix
cov_matrix = df_cleaned[numeric_cols].cov()
print("\n=== Covariance Matrix (first 5x5) ===")
print(cov_matrix.iloc[:5, :5])

# Optionally save cleaned & normalized data
df_cleaned.to_csv("DelayedFlights_Cleaned.csv", index=False)
print("\n✅ Cleaned & normalized dataset saved as 'DelayedFlights_Cleaned.csv'")
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler

# Load dataset
df = pd.read_csv("/content/drive/MyDrive/DelayedFlights.csv")

# Data Cleaning
df_cleaned = df.drop(columns=["Unnamed: 0"], errors="ignore")
df_cleaned = df_cleaned.dropna()

# Normalization
numeric_cols = df_cleaned.select_dtypes(include=[np.number]).columns
scaler = MinMaxScaler()
df_cleaned[numeric_cols] = scaler.fit_transform(df_cleaned[numeric_cols])

# Covariance Matrix
cov_matrix = df_cleaned[numeric_cols].cov()

# Plot Heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(cov_matrix, cmap="coolwarm", annot=False, cbar=True)
plt.title("Covariance Matrix Heatmap", fontsize=16)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("/content/drive/MyDrive/DelayedFlights.csv")

# --- Bar Graph Example: Flight counts per airline ---
plt.figure(figsize=(12,6))
df['UniqueCarrier'].value_counts().plot(kind='bar', color='skyblue')
plt.title("Number of Flights per Airline", fontsize=16)
plt.xlabel("Airline", fontsize=12)
plt.ylabel("Number of Flights", fontsize=12)
plt.show()





```

```python
# --- Bar Graph Example: Average arrival delay per airline ---
plt.figure(figsize=(12,6))
df.groupby("UniqueCarrier")["ArrDelay"].mean().sort_values().plot(kind="bar", color="orange")
plt.title("Average Arrival Delay per Airline", fontsize=16)
plt.xlabel("Airline", fontsize=12)
plt.ylabel("Average Arrival Delay (minutes)", fontsize=12)
plt.show()
```

```python
# --- Pie Chart Example: Cancelled vs Non-Cancelled Flights ---
cancel_counts = df["Cancelled"].value_counts()
plt.figure(figsize=(6,6))
cancel_counts.plot(kind="pie", autopct="%1.1f%%", labels=["Not Cancelled","Cancelled"], colors=["lightgreen","red"])
plt.title("Cancelled vs Non-Cancelled Flights", fontsize=14)
plt.ylabel("")  # remove default ylabel
plt.show()
```

```python
# --- Pie Chart Example: Flights per Day of Week ---
plt.figure(figsize=(7,7))
df["DayOfWeek"].value_counts().plot(kind="pie", autopct="%1.1f%%", cmap="Set3")
plt.title("Flights Distribution by Day of Week", fontsize=14)
plt.ylabel("")
plt.show()
```
